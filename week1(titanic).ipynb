```python
# 데이터 정의


import matplotlib.pyplot as plt

df = pd.DataFrame({
    'name':['Kim', 'Lee', 'Park', 'Choi', 'Hong', 'Chung','Jang'],
    'age':[22, 26, 78, 17, 46, 32, 21],
    'city':['Seoul', 'Busan', 'Seoul', 'Busan', 'Seoul', 'Daejun', 'Daejun'],
    'children':[2, 3, 0, 1, 3, 4, 3],
    'pets':[0, 1, 0, 2, 2, 0, 3]
})
```

```python
# 그래프


df.plot(kind='line', x='name', y='pets', color='red')
plt.show()
```

```python
# 중첩 차트 그리기
""" gca() <get current axis> == 현재 활성화된 축 ( axis ) <y축> 를 가져오는 함수
    ax == 그래프가 그려지는 캔버스 """


px = plt.gca()

df.plot(kind='line', x='name', y='children', ax=px)
df.plot(kind='line', x='name', y='pets', color='red', ax=px)

plt.show()
```

```python
# 타이타닉_사용 데이터


titanic = pd.read_csv('강의자료/1장/titanic.csv')
titanic.drop(['PassengerId', 'Ticket', 'Name'], inplace=True, axis=1)
titanic.head()
```

```python
# 타이타닉_피벗 테이블에서 인덱스를 사용하여 데이터를 그룹화하는 방법
""" pivot_table == 데이터프레임의 피벗 테이블을 생성하는 메서드
    데이터를 특정 열을 기준으로 그룹화하고, 그룹별로 집계 함수를 적용하여 
    새로운 형태의 테이블을 생성할 수 있음 """


table = pd.pivot_table(data=titanic, index=['Sex'])
print(table.dtypes)
```

```python
# 위의 피벗테이블 바 형태의 그래프로 그림
table.plot(kind="bar")
```

```python
# 타이타닉_다중 인덱스로 피벗하기


table = pd.pivot_table(titanic, index=['Sex', 'Pclass'])
```

```python
# 타이타닉_특징별로 다른 집계 함수 적용


table = pd.pivot_table(titanic, index=['Sex', 'Pclass'])
```

```python
# 타이타닉_value 매개 변수를 사용하여 특정한 데이터에 대한 집계
""" aggfunc 매개변수 == 피벗 테이블에서 사용할 집계 함수 ( aggregate function) 지정
    집계 함수 : 그룹화된 데이터를 요약하는데 사용 """


table = pd.pivot_table(titanic, index=['Sex', 'Pclass'],
                       values=['Survived'], aggfunc=np.mean)
```

```python
table.plot(kind='bar')
```

```python
# 타이타닉_데이터 간의 관계 찾기


table = pd.pivot_table(titanic, index=['Sex'],
                       columns=['Pclass'], values=["Survived"], aggfunc=np.sum)
```

```python
table.plot(kind='bar')
```
